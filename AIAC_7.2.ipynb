```python
def sum_upto_n(n):
  total = 0
  # The original code used range(1, n), which sums up to n-1.
  # To sum up to n, the range should be range(1, n+1).
  for i in range(1, n + 1):
    total += i
  return total

# Example usage:
# print(sum_upto_n(5)) # This should print 15 (1+2+3+4+5)
```

```python
class user:
  def __init__(self,name):
    self.name = name

  # AttributeError: The getname() method was called but not defined.
  # Adding a method to return the name.
  def getname(self):
    return self.name

u = user("Alice")
print(u.getname())
```

```python
def grade_student(score):
  # Logical Error: The original grading logic was incorrect for certain scores.
  # Correcting the logical error in grading:
  # score >= 90 for "A"
  # score < 70 for "B"
  # Otherwise "C"
  if score >= 90:
    return "A"
  elif score < 70:
    return "B"
  else:
    return "C"
```
