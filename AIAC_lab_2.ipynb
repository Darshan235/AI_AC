```python
import math

def calculate_circle_area(radius):
  """Calculates the area of a circle given its radius."""
  return math.pi * (radius ** 2)

def calculate_rectangle_area(length, width):
  """Calculates the area of a rectangle given its length and width."""
  return length * width
```

Here are examples of how to use the functions:

```python
# Example usage
radius = 5
circle_area = calculate_circle_area(radius)
print(f"The area of a circle with radius {radius} is: {circle_area}")

length = 10
width = 6
rectangle_area = calculate_rectangle_area(length, width)
print(f"The area of a rectangle with length {length} and width {width} is: {rectangle_area}")
```

## Tokenize the text

### Subtask:
Process the text to split it into individual words (tokens).


**Reasoning**:
Process the text to split it into individual words (tokens) by converting to lowercase, removing punctuation, splitting by whitespace, and handling empty strings.



```python
import string

# Convert to lowercase
text_content_lower = text_content.lower()

# Remove punctuation
text_content_no_punct = text_content_lower.translate(str.maketrans('', '', string.punctuation))

# Split into words and handle empty strings
words = [word for word in text_content_no_punct.split() if word]

print("Processed words:")
print(words)
```

## Count word frequencies

### Subtask:
Use a data structure (like a dictionary) to store and count the occurrences of each word.


**Reasoning**:
Create a dictionary to store word counts and iterate through the list of words to populate the dictionary.



```python
# Create an empty dictionary to store word frequencies
word_counts = {}

# Iterate through the list of words
for word in words:
  # For each word, check if it exists as a key in the dictionary.
  # If the word exists, increment its corresponding value (count).
  # If the word does not exist, add it as a new key to the dictionary with a value of 1.
  word_counts[word] = word_counts.get(word, 0) + 1

# Print the word counts
print("Word frequencies:")
print(word_counts)
```

### Subtask:
Print the words and their corresponding counts.


**Reasoning**:
Iterate through the word_counts dictionary and print each word and its count.



```python
# Iterate through the dictionary items
for word, count in word_counts.items():
  # Print the word and its count in a user-friendly format
  print(f"Word: {word}, Count: {count}")
```

```python
# Get input from the user
num_str = input("Enter a number: ")

# Convert the input to an integer (assuming valid integer input as exception handling is removed)
num = int(num_str)

# Calculate the number of digits
num_digits = len(num_str)

# Initialize sum of powers
sum_of_powers = 0

# Calculate the sum of the powers of its digits using a for loop
for digit_char in num_str:
  digit = int(digit_char)
  sum_of_powers += digit ** num_digits

# Check if it's an Armstrong number
if num == sum_of_powers:
  print(num, "is an Armstrong number.")
else:
  print(num, "is not an Armstrong number.")
```

```python
import string

def count_word_frequencies(file_path):
  """Reads a text file and counts the frequency of each word."""
  try:
    with open(file_path, 'r') as file:
      text_content = file.read()

    # Process the text
    text_content_lower = text_content.lower()
    text_content_no_punct = text_content_lower.translate(str.maketrans('', '', string.punctuation))
    words = [word for word in text_content_no_punct.split() if word]

    # Count word frequencies
    word_counts = {}
    for word in words:
      word_counts[word] = word_counts.get(word, 0) + 1

    return word_counts

  except FileNotFoundError:
    print(f"Error: The file '{file_path}' was not found.")
    return None
  except Exception as e:
    print(f"An error occurred: {e}")
    return None

# Example usage: Replace 'your_text_file.txt' with the path to your file
file_path = 'sample.txt'
word_frequencies = count_word_frequencies(file_path)

if word_frequencies:
  print(f"Word frequencies in '{file_path}':")
  for word, count in word_frequencies.items():
    print(f"Word: {word}, Count: {count}")
```

```python
import string
from collections import Counter

def count_word_frequencies_alternative(file_path):
  """Reads a text file and counts the frequency of each word using collections.Counter."""
  try:
    with open(file_path, 'r') as file:
      text_content = file.read()

    # Process the text
    text_content_lower = text_content.lower()
    text_content_no_punct = text_content_lower.translate(str.maketrans('', '', string.punctuation))
    words = [word for word in text_content_no_punct.split() if word]

    # Count word frequencies using Counter
    word_counts = Counter(words)

    return word_counts

  except FileNotFoundError:
    print(f"Error: The file '{file_path}' was not found.")
    return None
  except Exception as e:
    print(f"An error occurred: {e}")
    return None

# Example usage: Replace 'your_text_file.txt' with the path to your file
file_path = 'sample.txt'
word_frequencies = count_word_frequencies_alternative(file_path)

if word_frequencies:
  print(f"Word frequencies in '{file_path}' (using Counter):")
  for word, count in word_frequencies.items():
    print(f"Word: {word}, Count: {count}")
```

```python
# Example list of dictionaries
list_of_dictionaries = [
    {'name': 'Alice', 'age': 30},
    {'name': 'Bob', 'age': 25},
    {'name': 'Charlie', 'age': 35},
    {'name': 'David', 'age': 25}
]

# Specify the key to sort by
sort_key = 'age'

# Sort the list of dictionaries by the specified key
sorted_list = sorted(list_of_dictionaries, key=lambda x: x[sort_key])

# Print the sorted list
print(f"Original list: {list_of_dictionaries}")
print(f"Sorted list by '{sort_key}': {sorted_list}")

# You can also sort in reverse order
sorted_list_reverse = sorted(list_of_dictionaries, key=lambda x: x[sort_key], reverse=True)
print(f"Sorted list by '{sort_key}' in reverse: {sorted_list_reverse}")
```
