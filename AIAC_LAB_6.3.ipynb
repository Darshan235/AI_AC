```python
class Student:
    def __init__(self, name, age, major):
        self.name = name
        self.age = age
        self.major = major

    def display_details(self):
        print(f"Name: {self.name}")
        print(f"Age: {self.age}")
        print(f"Major: {self.major}")

# Take input dynamically from the user
name = input("Enter student name: ")
age = int(input("Enter student age: "))  # Assuming age is an integer
major = input("Enter student major: ")

# Create an instance of the Student class with dynamic input
student1 = Student(name, age, major)

# Display the student's details
student1.display_details()
```

```python
# Take multiple ages input dynamically from the user
ages_input = input("Enter ages separated by commas (e.g., 5, 15, 25): ")
ages = [int(age.strip()) for age in ages_input.split(',')]

# Classify age groups for each age
for age in ages:
    if 0 <= age <= 2:
        print(f"{age}: Infant")
    elif 3 <= age <= 12:
        print(f"{age}: Child")
    elif 13 <= age <= 17:
        print(f"{age}: Teenager")
    elif 18 <= age <= 19:
        print(f"{age}: Adult Teenager")
    elif 20 <= age <= 39:
        print(f"{age}: Adult")
    elif 40 <= age <= 59:
        print(f"{age}: Middle-aged Adult")
    elif 60 <= age <= 80:
        print(f"{age}: Senior Adult")
    elif 81 <= age <= 150:
        print(f"{age}: Elderly")
    else:
        print(f"{age}: Age group not classified in this program.")
```

```python
def sum_to_n(n):
  """Calculates the sum of the first n numbers using a while loop."""
  total_sum = 0
  i = 1
  while i <= n:
    total_sum += i
    i += 1
  return total_sum

# Take input dynamically from the user
num = int(input("Enter a number (n) to calculate the sum of the first n numbers: "))

# Calculate and print the sum
result = sum_to_n(num)
print(f"The sum of the first {num} numbers is: {result}")
```

```python
class BankAccount:
    def __init__(self, account_holder, initial_balance=0):
        self.account_holder = account_holder
        self.balance = initial_balance

    def deposit(self, amount):
        if amount > 0:
            self.balance += amount
            print(f"Deposited ${amount}. New balance: ${self.balance}")
        else:
            print("Deposit amount must be positive.")

    def withdraw(self, amount):
        if amount > 0:
            if self.balance >= amount:
                self.balance -= amount
                print(f"Withdrew ${amount}. New balance: ${self.balance}")
            else:
                print("Insufficient funds.")
        else:
            print("Withdrawal amount must be positive.")

    def get_balance(self):
        return self.balance

# Take input dynamically from the user
account_holder_name = input("Enter account holder name: ")
initial_deposit = float(input("Enter initial deposit amount: "))

# Create a bank account instance
my_account = BankAccount(account_holder_name, initial_deposit)

# Perform some transactions (example)
my_account.deposit(float(input("Enter deposit amount: ")))
my_account.withdraw(float(input("Enter withdrawal amount: ")))

# Display the final balance
print(f"Final balance: ${my_account.get_balance()}")
```
